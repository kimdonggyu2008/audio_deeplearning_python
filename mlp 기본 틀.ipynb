```python
import numpy as np
```

```python
class MLP(object):
#3개 입력을 받고 2개의 출력을 시행함
  def __init__(self,num_inputs=3,hidden_layers=[3,3], num_outputs=2):
    #입력 갯수, 레이어 갯수, 출력 데이터 갯수 지정

    self.num_inputs=num_inputs
    self.hidden_layers=hidden_layers# 첫번째 층이 3개 뉴런, 두번째 층이 3개 뉴런을 가지게 만듦
    self.outputs=num_outputs

    layers = [num_inputs] + hidden_layers + [num_outputs]

    self.weights=[] #랜덤한 가중치용 배열 생성
    for i in range(len(layers)-1):
      w=np.random.rand(layers[i],layers[i+1])#i번째 레이어 -> i+1번째 레이어로 넘어가는 과정의 가중치 갯수 세기
      self.weights.append(w)# 다음 층으로 넘어갈때 각 레이어별 갯수가 맞아야 함

  def forward_propagate(self,inputs):
    activations=inputs#활성화 함수
    for w in self.weights:
      net_inputs=np.dot(activations,w)#행렬곱
      activations=self._sigmoid(net_inputs)#주어진 행렬곱의 시그모이드로 구별시킴

    return activations

  def _sigmoid(self,x):#시그모이드 함수
    y=1.0/(1+np.exp(-x))
    return y


if __name__=="__main__":
  mlp=MLP()

  inputs=np.random.rand(mlp.num_inputs)
  print(inputs)
  output=mlp.forward_propagate(inputs)
  print(output)
```
