```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
import numpy as np
from random import random

#데이터 셋 생성
def generate_dataset(num_samples,test_size=0.33):
  x=np.array([[random()/2 for _ in range(2)] for _ in range(num_samples)])
  y=np.array([i[0]+i[1] for i in x])

  x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=test_size)
  return x_train,x_test,y_train,y_test



#메인 실행
if __name__=="__main__":

  x_train, x_test, y_train, y_test = generate_dataset(5000, 0.3)
#모델 구성
  model=tf.keras.models.Sequential([
      tf.keras.layers.Dense(5,input_dim=2,activation="sigmoid"),#뉴런 5개,완전연결 모델, 시그모이드 활성함수 지정
      tf.keras.layers.Dense(1,activation="sigmoid")#뉴런 1개
  ])
#최적화함수, 최솟값 찾아가는 함수 지정
  optimizer=tf.keras.optimizers.SGD(learning_rate=0.1)
#모델 구축
  model.compile(optimizer=optimizer,loss='mse')# 평균제곱오차 구함
#가중치 계산
  model.fit(x_train,y_train,epochs=100)
#각 계산별로 결과 출력, verbose값에 따라 달라짐, 0이면 아무것도 안나옴, 1이면 진행바만, 2면 에포크마다 막대, 결과 나옴
  model.evaluate(x_test,y_test,verbose=2)
#주어진 값들의 합을 예측하는 모델 구축
  data = np.array([[0.1, 0.2], [0.2, 0.2]])
#예측 시행
  predictions = model.predict(data)

    # print predictions
  print("\nPredictions:")
  for d, p in zip(data, predictions):
    print("{} + {} = {}".format(d[0], d[1], p[0]))
```
